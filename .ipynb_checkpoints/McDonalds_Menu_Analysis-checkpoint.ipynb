# Question 1
## How many calories does the average McDonald's value meal contain? 
---
A value meal from McDonalds is any one of their main items (hamburgers, fish, chicken), with a medium french fry or salad and a drink.

```python
# Check the categories to see where the information we want would be.
# Could just print them, but I saved the list to a variable just in case we need them later.
categories = menu['Category'].unique()
print(categories)
```

```python
# Create a new df with only the categories that matter for us here. In this case the category, item name, and calorie number.
menu_calories = menu[['Category', 'Item', 'Calories']]
menu_calories.head(-5)
```

```python
# Pull out the snacks and sides. For this however, we only want Medium French Fries. we'll pull that out individually later.
Snacks_and_sides = menu_calories[menu_calories['Category'] == 'Snacks & Sides'].copy()
Snacks_and_sides.head()
```

```python
# Pull out the sandwiches (though this does include chicken nuggets too) into their own dataframe.
sandwiches = menu_calories[(menu_calories['Category'].isin(['Beef & Pork', 'Chicken & Fish']))].copy()
sandwiches.head()
```

```python
# Pull out the salads into their own data frame.
salads = Snacks_and_sides = menu_calories[menu_calories['Category'] == 'Salads'].copy()
salads
```

```python
# Pull out the drinks into their own dataframe.
drinks = menu_calories[menu_calories['Category'] == 'Beverages'].copy()
```

```python
# Use merge to create the cartesian product of the different menu items
# to calculate the total caloric value of each kind of meal.
from pandas import merge

# Add a 'key' column and set it to 1 so that we can merge dataframes together.
drinks['key'] = 1
salads['key'] = 1
sandwiches['key'] = 1

# Create cartesian product of drinks and salads.
drink_salad_cartesian = merge(drinks, salads, on='key')

# Create cartesian product of drinks, salads, and sandwiches.
sandwich_drink_salad = merge(drink_salad_cartesian, sandwiches, on='key')

# Check to make sure it worked.
sandwich_drink_salad.head()
```

```python
# Rename the columns so that they make more sense.
sandwich_drink_salad.rename(columns = {'Calories_x':'Beverage_Calories',
                                      'Calories_y': 'Salad_Calories',
                                      'Calories': 'Main_Dish_Calories'}, inplace = True)
sandwich_drink_salad.drop(['key'], axis = 1)
```

```python
# Create a new column that is the sum of calories for each row
sandwich_drink_salad['Total Calories'] = sandwich_drink_salad['Beverage_Calories'] + sandwich_drink_salad['Salad_Calories'] + sandwich_drink_salad['Main_Dish_Calories']
```

```python
sandwich_drink_salad.head()
```

```python
# Print the average calorie count for a value meal
# Note: This does not count fries yet.
print(f'The average caloric count in a value meal is:', sandwich_drink_salad['Total Calories'].mean().round())
```

```python
# Merge together drinks and sandwiches to a new dataframe
drinks_sandwiches = merge(drinks, sandwiches, on='key')
```

```python
# Pull out just the snacks and sides columns
Snacks_and_sides = menu_calories[menu_calories['Category'] == 'Snacks & Sides'].copy()
Snacks_and_sides.head()
```

```python
# Pull out the medium french fries
med_french_fries = menu_calories[menu_calories['Item'] == 'Medium French Fries'].copy()
med_french_fries
```

```python
# Add a key value of 1 to the french fries so that we can perform a cartesian product
med_french_fries['key'] = 1
fries_drinks_sandwiches = merge(drinks_sandwiches, med_french_fries, on='key')
```

```python
fries_drinks_sandwiches.head()
```

```python
# Rename some of the columns so that they make more sense
fries_drinks_sandwiches.rename(columns = {'Calories_x':'Beverage_Calories',
                                      'Calories_y': 'Sandwich_Calories',
                                      'Calories': 'French_Fries_Calories'}, inplace = True)

# Create a new column called "total calories" based on the sum of all the different calorie columns
fries_drinks_sandwiches['Total Calories'] = fries_drinks_sandwiches['Beverage_Calories'] + fries_drinks_sandwiches['Sandwich_Calories'] + fries_drinks_sandwiches['French_Fries_Calories']
```

```python
fries_drinks_sandwiches.head()
```

```python
# Append the two value menu dataframes together to get one dataframe with all of the options.
all_value_menu_options = sandwich_drink_salad.append(fries_drinks_sandwiches)
```

```python
# Note that we now have NaN values. This is due to our renaming of columns earlier. In this case it doesn't matter, since we have the same 'Total Calories'
# column that we will be using.
all_value_menu_options.head()
```

```python
average_calories_value_menu = all_value_menu_options['Total Calories'].mean()
```

```python
print(f'The average caloric value of the McDonalds value menu is: {average_calories_value_menu.round()}.')
```

# Question 2
## How much do beverages, like soda or coffee, contribute to the overall caloric intake?

```python
# Calculate the average number of calories drinks add to a McDonalds meal.
average_drink_calories = drinks['Calories'].mean()
print(f'On average, drinks from McDonalds add {average_drink_calories.round()} calories to a meal.')
```

However, we can see in the list that there are a lot of diet drinks that add 0 calories. We don't have meal information to see how popular diet drinks are over regular drinks, but we should still try and remove these to see how much it was affecting the average.

```python
# Remove the drinks that do not have any calories, as this can pull the average down. Again, we do not have numbers on how often different drinks are purchase,
# so this could give us weird results. Just something to keep in mind.
drink_calories_without_diet_or_water = drinks[drinks['Calories'] != 0]
drink_calories_without_diet_or_water.head()
```

```python
# Calculate the average
avg_drink_calories_without_diet_or_water = drink_calories_without_diet_or_water['Calories'].mean()
print(f'The average caloric increase from drinks at McDonalds without diet soda or water is {avg_drink_calories_without_diet_or_water.round()}.')
```

We can see that the difference in calories is 171 - 114 = 57 calories added on average for non-diet and non-water drinks.

# Q4
## What about ordering egg whites instead of whole eggs?

```python
egg_items = menu[menu['Item'].str.contains("Egg|Eggs")]
egg_items.head()
```

```python
egg_whites_dishes = egg_items[egg_items['Item'].str.contains("Whites")]
avg_egg_white_calories = egg_whites_dishes['Calories'].mean()
print(avg_egg_white_calories.round())
```

```python
non_egg_whites_dishes = egg_items[~egg_items['Item'].str.contains("Whites")]
avg_non_egg_white_calories = non_egg_whites_dishes['Calories'].mean()
print(avg_non_egg_white_calories.round())
```

```python
print(f'Average nutrional value from dishes with egg whites:\n\n{egg_whites_dishes.mean().round(2)}')
print('\n')
print(f'Average nutrional value from dishes WITHOUT egg whites:\n\n{non_egg_whites_dishes.mean().round(2)}')
```

Here we can see that dishes without egg whites typically have lower values across the board.

# Data Visualization
While looking at numbers works for some people, it can be important to visualize data as well.

```python
# Createa  scatterplot of Total Fat vs Calories and color it by category.
fig = pplot.gcf()
fig.set_size_inches(12, 8)
scatter = sns.scatterplot(data = menu, x = 'Calories', y = 'Total Fat', hue = 'Category', size = 30)
```

We can see that there is an outlier on the upper right corner that makes this plot hard to glean information from.

```python
# Let's try creating a new plot but with limits on the x and y axis to 'zoom in' on the data.
fig = pplot.gcf()
fig.set_size_inches(12, 8)
scatter = sns.scatterplot(data = menu, x = 'Calories', y = 'Total Fat', hue = 'Category')
scatter.set_xlim(0, 1250)
scatter.set_ylim(0, 65)
```

This is better, but it still is hard to look at due to how many categories we have.

```python
grid = sns.FacetGrid(menu, col = "Category", hue = "Category", col_wrap=5)
grid.map(sns.scatterplot, "Total Fat", "Calories")

grid.add_legend()
pplot.show()
```

Again, we see that the outlier is preventing us from seeing the information that we need. Let's cap the x_axis again.

```python
# Use FaceGrid in seaborn to break up the graphs by category. This lets us see a side by side by side view of all of the different foods.
grid = sns.FacetGrid(menu, col = "Category", hue = "Category", col_wrap=5)
grid.map(sns.scatterplot, "Calories", "Total Fat")

# Set the x axis limit to 1250 to trim the graphs down. This will only cut off the outlier in the 'Chicken and Fish' category.
grid.set(xlim = (0, 1250))

grid.add_legend()
pplot.show()
```
